```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../../../Divar.csv', index_col=0)
```

```python
tmp = pd.DataFrame(df['rent_value'])
tmp['rent_value'] = tmp['rent_value'] + 0.000001
tmp.info()
```

```python
len(tmp[tmp['rent_value'] == 0.000001])
```

```python
def cap_outliers_iqr(df, column):
    """
    جایگزینی outlierها با حدود بالا و پایین بر اساس IQR
    """
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # کپی از دیتافریم
    df_capped = df.copy()

    # جایگزینی مقادیر خارج از محدوده
    df_capped.loc[df_capped[column] < lower_bound, column] = lower_bound
    df_capped.loc[df_capped[column] > upper_bound, column] = upper_bound

    return df_capped
```

```python
tmp1 = cap_outliers_iqr(tmp,'rent_value')
```

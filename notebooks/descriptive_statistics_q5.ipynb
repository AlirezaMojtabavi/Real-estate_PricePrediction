```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/mnt/e/bootcamp/Project/azmmon_farz/utm_coords1.csv")
```

```python
df.info()
```

```python
#   Heatmap  hexbin
plt.figure(figsize=(10, 5.5), facecolor='white')  #  

hb = plt.hexbin(
    x=df['UTM_Easting'],
    y=df['UTM_Northing'],
    gridsize=250,       #    
    cmap='hot',
    linewidths=0,
    mincnt=1            #         
)

plt.colorbar(hb, label='Density')  #     
plt.xlabel("UTM Easting (m)")
plt.ylabel("UTM Northing (m)")
plt.title("Heatmap of UTM Points (Hexbin with 'hot' colormap)")
plt.grid(False)
plt.tight_layout()
plt.show()
```

### Calculating Unit_price for Prediction !!!

براساس مقادیر مختلف ستون <br>
cat3_slug <br>
محاسبه ستون قیمت واحد متفاوت است <br>
اگر در داده ما مقدار این ستون ار مقادیر زیر باشد <br>
house-villa-sell, house-villa-rent, plot-old ---- > land_size <br>
else ----> building_size <br>
ردیف های مربوط به مشارکتی هم که حذف شده اند

```python
df["final_size"] = np.where(
    df["cat3_slug"].isin(["house-villa-sell", "house-villa-rent", "plot-old"]),
    df["land_size"],
    df["building_size"]
)

df["unit_price"] = df["total_price"] / df["final_size"]
```

### Creating model for estimate unit_price

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_val_score
```

```python
numeric_features = ["floor", "total_floors_count", "unit_per_floor", "final_size"]
X = df[numeric_features]
y = df["unit_price"]


imputer = SimpleImputer(strategy="median")
X_imputed = imputer.fit_transform(X)

# تقسیم داده‌ها
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestRegressor(
    n_estimators=5,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
```

```python
print("R² score:", r2_score(y_test, y_pred))
print("RMSE:", mean_squared_error(y_test, y_pred))
print(accuracy_score()
```

```python
scores = cross_val_score(rf_model, X_imputed, y, cv=5, scoring='r2')
print(scores)
```

### simple regression

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score


numeric_features = ["floor", "total_floors_count", "unit_per_floor", "final_size"]
X = df[numeric_features]
y = df["unit_price"]

imputer = SimpleImputer(strategy="median")
X_imputed = imputer.fit_transform(X)

# نرمال‌سازی ویژگی‌ها
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# مدل
lin_reg = LinearRegression()

# Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
r2_scores, rmse_scores = [], []

for train_index, test_index in kf.split(X_scaled):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    lin_reg.fit(X_train, y_train)
    y_pred = lin_reg.predict(X_test)

    r2_scores.append(r2_score(y_test, y_pred))
    rmse_scores.append(np.sqrt(mean_squared_error(y_test, y_pred)))

print("R² scores:", r2_scores)
print("Mean R²:", np.mean(r2_scores))
print("RMSE scores:", rmse_scores)
print("Mean RMSE:", np.mean(rmse_scores))
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# --- 1. آماده‌سازی داده ---
numeric_features = ["floor", "total_floors_count", "unit_per_floor", "final_size"]
X = df[numeric_features]
y = np.log1p(df["unit_price"])   # لگاریتم هدف

# --- 2. پر کردن NaN با میانه ---
imputer = SimpleImputer(strategy="median")
X_imputed = imputer.fit_transform(X)

# --- 3. استانداردسازی ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# --- 4. مدل ---
lin_reg = LinearRegression()

# --- 5. Cross-Validation ---
kf = KFold(n_splits=5, shuffle=True, random_state=42)
r2_scores, rmse_scores = [], []

for train_index, test_index in kf.split(X_scaled):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    lin_reg.fit(X_train, y_train)
    y_pred_log = lin_reg.predict(X_test)

    # برگردوندن پیش‌بینی‌ها به مقیاس اصلی
    y_pred = np.expm1(y_pred_log)
    y_true = np.expm1(y_test)

    r2_scores.append(r2_score(y_true, y_pred))
    rmse_scores.append(np.sqrt(mean_squared_error(y_true, y_pred)))

print("R² scores:", r2_scores)
print("Mean R²:", np.mean(r2_scores))
print("RMSE scores:", rmse_scores)
print("Mean RMSE:", np.mean(rmse_scores))
```

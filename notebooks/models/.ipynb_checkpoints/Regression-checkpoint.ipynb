### Load Neccessary Data

```python
sect1  = pd.read_csv('cleaned_0_8.csv')
sect1.head()
sect1 = sect1.drop(['Unnamed: 0.1', 'Unnamed: 0', 'UTM_Easting', 'UTM_Northing','neighborhood_slug'], axis=1)
sect1
```

```python
sect2 = pd.read_csv('cleaned_23_42.csv')
sect2 = sect2[['land_size','building_size','floor','total_floors_count','unit_per_floor','has_balcony']]
sect2
```

```python
totalprice = pd.read_csv('../../processed data/total_price.csv') 
totalprice
```

### merging all of sections

```python
df = pd.concat([sect1,sect2,totalprice['total_price']], axis=1)
#df = df.drop('Unnamed: 0',axis=1)
df
```

```python
df.info()
```

ردیف هایی که ستون توتال پرایسشون مقدار نداره رو دراپ میکنیم
علاوه بر اون ردیف هایی که پارمتر نوع قراردشون <br>
cat3_slug = 'partnership' <br>
رو هم بر اساس بررسی و فقدان داده در این نوع دسته بندی دراپ میکنیم

```python
p1 = df['has_balcony'].mean()  # احتمال 1 بودن
missing_idx = df['has_balcony'].isna()

df.loc[missing_idx, 'has_balcony'] = np.random.binomial(1, p1, size=missing_idx.sum())
```

```python
def fill_total_price_by_group(df):
    
    df = df.copy()
    df["total_price"] = df.groupby(["cat3_slug", "city_slug"])["total_price"]\
                          .transform(lambda x: x.fillna(x.mean()))
    return df
```

```python
df_filled = fill_total_price_by_group(df)

print("before:", df["total_price"].isna().sum())
print("after:", df_filled["total_price"].isna().sum())
```

```python
df_filled.info()
```

ستون های مربوط به متراژ <br>
land_size , building_size <br>
ازشون لگاریم گرفته شده بود که مبنای خودشون باید برگردند

```python
# df_filled["building_size"] = np.exp(df_filled["building_size"])
# df_filled["land_size"] = np.exp(df_filled["land_size"])
df_filled['total_price_log'] = np.log(df_filled["total_price"]) / np.log(6)
```

### Calculating Unit_price for Prediction !!!

براساس مقادیر مختلف ستون <br>
cat3_slug <br>
محاسبه ستون قیمت واحد متفاوت است <br>
اگر در داده ما مقدار این ستون ار مقادیر زیر باشد <br>
house-villa-sell, house-villa-rent, plot-old ---- > land_size <br>
else ----> building_size <br>
ردیف های مربوط به مشارکتی هم که حذف شده اند

```python
df_filled["final_size"] = np.where(
    df_filled["cat3_slug"].isin(["house-villa-sell", "house-villa-rent", "plot-old"]),
    df_filled["land_size"],
    df_filled["building_size"]
)

df_filled["unit_price"] = df_filled["total_price_log"] / df_filled["final_size"]
```

### Creating model for estimate unit_price

#### Random forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_squared_error
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
```

#### Data preperation

```python
df_0 = df_0[~np.isinf(df_0["unit_price"]) & df_0["unit_price"].notna()]
le = LabelEncoder()
# آموزش و تبدیل
df_0['city_slug'] = le.fit_transform(df_0['city_slug'].astype(str))
df_0['cat3_slug'] = le.fit_transform(df_0['cat3_slug'].astype(str))
df_0
```

```python
numeric_features = ['cat3_slug','city_slug',"floor", "total_floors_count", "unit_per_floor"]
X = df_0[numeric_features]
y = df_0["unit_price"]


imputer = SimpleImputer(strategy="median")
X_imputed = imputer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestRegressor(
    n_estimators=25,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("R2 Score:", r2)
print("MAE:", mae)
print("MSE:", mse)
```

```python
scores = cross_val_score(rf_model, X_imputed, y, cv=5, scoring='r2')
print(scores)
```

### simple regression

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score


imputer = SimpleImputer(strategy="median")
X_imputed = imputer.fit_transform(X)

# نرمال‌سازی ویژگی‌ها
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# مدل
lin_reg = LinearRegression()

# Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
r2_scores, rmse_scores = [], []

for train_index, test_index in kf.split(X_scaled):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    lin_reg.fit(X_train, y_train)
    y_pred = lin_reg.predict(X_test)

    r2_scores.append(r2_score(y_test, y_pred))
    rmse_scores.append(np.sqrt(mean_squared_error(y_test, y_pred)))

print("R² scores:", r2_scores)
print("Mean R²:", np.mean(r2_scores))
print("RMSE scores:", rmse_scores)
print("Mean RMSE:", np.mean(rmse_scores))
```
